```python
import json
def load_result(file):
    res = []
    with open(file, "r", encoding = "utf-8") as fp:
        for line in fp:
            d = json.loads(line.strip())
            res.append(d)
    return res
```

```python
import json

#准确率
def get_acc(dataset):
    correct = 0
    total = 0

    for item in dataset:
        # 获取chosen和rejected的分数
        score_chosen = item["chosen"]
        score_rejected = item["rejected"]

        # 统计正确率
        if score_chosen > score_rejected:
            correct += 1
        total += 1

    accuracy = correct / total * 100
    print(f"奖励模型准确率：{accuracy:.2f}%")
    return accuracy

#auc计算
from sklearn.metrics import roc_auc_score
def calculate_auc(dataset):
    y_true = []
    y_score = []
    
    for item in dataset:
        score_chosen = item["chosen"]
        score_rejected = item["rejected"]
        
        # 分数差异作为预测置信度
        y_score.append(score_chosen - score_rejected)
        y_true.append(1)  # chosen应该比rejected得分高
        
        # 添加反向样本（可选，增强鲁棒性）
        y_score.append(score_rejected - score_chosen)
        y_true.append(0)
    
    return roc_auc_score(y_true, y_score)

from scipy.stats import kendalltau

#Kendall Tau 相关系数
def calculate_kendall_tau(dataset):
    model_rankings = []
    human_rankings = []
    
    for item in dataset:
        # 获取模型评分
        scores = {
            "chosen": item["chosen"],
            "rejected": item["rejected"]
        }
        
        # 人类标注的排序（chosen > rejected 记为 [1, 0]）
        human_rank = [1, 0]
        
        # 模型的排序（按分数降序）
        model_rank = sorted(
            ["chosen", "rejected"], 
            key=lambda x: -scores[x]
        ).index("chosen")  # 返回chosen的排名（0或1）
        
        model_rankings.append(model_rank)
        human_rankings.append(human_rank[0])  # 人类标注中chosen总是排第一
    print(model_rankings, human_rankings)
    res = kendalltau(model_rankings, human_rankings)
    tau = res.correlation
    return tau
```
